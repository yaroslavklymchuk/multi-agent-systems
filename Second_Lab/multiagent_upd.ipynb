```python
import numpy as np

from visualize import DrawChessPosition
from figures import (King, Pawn, Queen, Knight, Rook, Bishop)
from board import ChessBoard
from algorithms import DFS, ABT
```

### Generate Random Chess Board for DFS

```python
CONFIG = {'B': 3, 'N': 2, 'P': 4, 'R': 1, 'Q': 1}
SIZE = 10
```

```python
chess_board = ChessBoard(SIZE, CONFIG)
```

```python
board_generated, gen = chess_board.generate_board()
```

```python
board_generated
```

```python
board_matrix, board_symbols_words, board_mapping, figures_mapping = chess_board.get_coordinates_map(
    board_generated, gen)
```

## DFS

```python
figures_list = list(figures_mapping.values())
```

```python
empty_board = np.zeros((SIZE, SIZE)).astype(str)
```

```python
dfs = DFS(empty_board, figures_list)
```

```python
generated_board_fen_dfs = dfs.main()
```

```python
generated_board_fen_dfs
```

```python
generated_board_fen_dfs_render = DrawChessPosition(n=SIZE)
generated_board_dfs = generated_board_fen_dfs_render.draw(generated_board_fen_dfs)
```

```python
generated_board_dfs
```
